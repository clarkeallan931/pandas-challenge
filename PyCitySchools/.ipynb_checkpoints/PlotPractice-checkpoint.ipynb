```python
import matplotlib.pyplot as plt
users = [13000,26000,52000,30000,9000]
x_axis = np.arange(len(users))
plt.bar(x_axis, users, color='r', alpha=0.5, align="center")
plt.xticks(x_axis, ["Java", "C++", "Python", "Ruby", "Clojure"])
plt.show()
```

```python
cities = ["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"]
cars_in_cities = [214.7, 564.4, 416.5, 466.7, 350.6]
x_axis = np.arange(len(cars_in_cities))
plt.bar(x_axis, cars_in_cities, color='b', alpha=0.5, align="center")
plt.xticks(x_axis, cities)
plt.title("Density of Commuting Cars in Cities")
plt.xlabel("Cities")
plt.ylabel("Communting Cares Per 1,000 Population Age 16+")
plt.savefig("CarDensity.png")
plt.show()
```
